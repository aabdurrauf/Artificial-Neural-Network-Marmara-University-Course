```python
# add noise (randomly adding black, grey, or white pixel)
def add_noise(m, fonts):
  for font in fonts:
    for i in range(m):
      x = random.randint(0, 6)
      y = random.randint(0, 8)
      noise = random.choice([-1, 0, 1])
      font[y, x] = noise
  return fonts
```

```python
# add noise (randomly select a pixel and reverse the color)
def add_noise2(m, fonts):
  for font in fonts:
    for i in range(m):
      x = random.randint(0, 6)
      y = random.randint(0, 8)
      font[y, x] *= -1
  return fonts
```

```python
# generating new datasets with noises
def iterate_noise(n, m, input):
  for i in range(n):
    font = np.copy(input)
    if i == 0:
      # font_noise = np.array(add_noise(m, font))
      font_noise = np.array(add_noise2(m, font))
    else:
      # font_noise = np.append(font_noise, add_noise(m, font), axis=0)
      font_noise = np.append(font_noise, add_noise2(m, font), axis=0)
  return font_noise
```

```python
# generate datasets
inputs1_noise = iterate_noise(100, 7, inputs1)
inputs2_noise = iterate_noise(100, 7, inputs2)
inputs3_noise = iterate_noise(100, 7, inputs3)
```

```python
# seperating the training input datasets (70% training and 30% testing)
train_dataset1 = np.append(inputs1, inputs1_noise[:488], axis=0)
train_dataset2 = np.append(inputs2, inputs2_noise[:488], axis=0)
train_dataset3 = np.append(inputs3, inputs3_noise[:488], axis=0)

train_dataset_input = np.vstack((train_dataset1, train_dataset2, train_dataset3))
```

```python
# seperating the testing input datasets (70% training and 30% testing)
test_dataset1 = np.array(inputs1_noise[488:])
test_dataset2 = np.array(inputs2_noise[488:])
test_dataset3 = np.array(inputs3_noise[488:])

test_dataset_input = np.vstack((test_dataset1, test_dataset2, test_dataset3))
```

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten
```

***Set the model name, activation function, and learning rate hear***

```python
# SETTING THE PARAMETER
NUM = "rsprop_#2_06"

ACTIVATION_FUNCTION = "relu"
# ACTIVATION_FUNCTION = "sigmoid"
# ACTIVATION_FUNCTION = "tanh"

# LR = 0.001
LR = 0.01
# LR = 0.2
```

```python
# build the model
model=Sequential([
    Flatten(input_shape=(9,7,1)),
    Dense(10,activation=ACTIVATION_FUNCTION),
    Dense(7,activation="softmax")
    ])
```

```python
model.summary()
```

***To select an optimizer algorithm for training the model, uncommant a code block and comment the others***

```python
# select the optimizer
# import keras
# opt = keras.optimizers.Adam(learning_rate=LR)
# model.compile(optimizer=opt,loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
import keras
opt = keras.optimizers.RMSprop(learning_rate=LR)
model.compile(optimizer=opt,loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
# import keras
# opt = keras.optimizers.SGD(learning_rate=LR)
# model.compile(optimizer=opt,loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
# training the model
history = model.fit(train_dataset_input,train_dataset_output,epochs=11,batch_size=1,validation_split=0.2)
```

```python
# saving the model
model_name = "model_" + NUM
model.save(model_name)
```

```python
# plotting the training history
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.figure()
plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
```

```python
# testing the model
outputs_predict=model.predict(test_dataset_input)
```
